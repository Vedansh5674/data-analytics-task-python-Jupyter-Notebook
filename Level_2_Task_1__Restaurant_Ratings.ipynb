```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/Dataset .csv")

# -----------------------------
# Task 1: Distribution of Ratings
# -----------------------------

# Plot histogram of aggregate ratings
plt.figure(figsize=(8,5))
plt.hist(df['Aggregate rating'], bins=10)
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Aggregate Ratings")
plt.show()

# Most common rating range
rating_range = df['Aggregate rating'].value_counts().idxmax()

print("Most Common Rating:", rating_range)

# -----------------------------
# Task 2: Average Number of Votes
# -----------------------------

average_votes = df['Votes'].mean()

print("Average Number of Votes:", average_votes)
```

```python
# Rating distribution count
rating_distribution = df['Aggregate rating'].value_counts().sort_index()

print("\nRating Distribution:")
print(rating_distribution)

# Average votes rounded
print("\nAverage Number of Votes:",
      round(average_votes, 2))
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Aggregate rating'], bins=10, edgecolor='black')
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Aggregate Ratings")
plt.grid(True)
plt.show()
```
