```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
import pandas as pd
df = pd.read_csv("Dataset .csv")
```

```python
# Check 'Has Online delivery' Column
print(df['Has Online delivery'].head())
```

```python
# Calculate Percentage of Restaurants Offering Online Delivery
delivery_counts = df['Has Online delivery'].value_counts()

percentage = (delivery_counts / len(df)) * 100

print(percentage)
```

```python
for i, value in percentage.items():
    print(f"{i}: {value:.2f}%")
```

```python
# Create Bar Chart
plt.bar(delivery_counts.index, delivery_counts.values)

plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.title("Restaurants Offering Online Delivery")

plt.show()
```

```python
# Compare Average Ratings
avg_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print(avg_rating)
```

```python
# Bar Chart for Average Ratings
plt.bar(avg_rating.index, avg_rating.values)

plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.title("Average Ratings: Online Delivery vs No Delivery")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Dataset .csv")

# Online delivery percentage
delivery_counts = df['Has Online delivery'].value_counts()

percentage = (delivery_counts / len(df)) * 100

for i, value in percentage.items():
    print(f"{i}: {value:.2f}%")

# Bar chart
plt.bar(delivery_counts.index, delivery_counts.values)

plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.title("Restaurants Offering Online Delivery")

plt.show()

# Average rating comparison
avg_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print(avg_rating)

# Rating comparison chart
plt.bar(avg_rating.index, avg_rating.values)

plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.title("Average Ratings Comparison")

plt.show()
```

```python
# Pie Chart for Online Delivery
plt.pie(delivery_counts.values,
        labels=delivery_counts.index,
        autopct='%1.1f%%')

plt.title("Online Delivery Availability")

plt.show()
```

```python
# Add Values on Bar Chart
bars = plt.bar(delivery_counts.index, delivery_counts.values)

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, y,
             ha='center', va='bottom')

plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.title("Restaurants Offering Online Delivery")

plt.show()
```
