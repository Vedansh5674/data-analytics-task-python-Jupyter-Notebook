```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/Dataset .csv")

# -----------------------------
# Task 1: Most Common Cuisine Combinations
# -----------------------------

# Count cuisine combinations
cuisine_counts = df['Cuisines'].value_counts()

print("Most Common Cuisine Combinations:")
print(cuisine_counts.head(10))

# Bar chart of top 10 cuisine combinations
top_cuisines = cuisine_counts.head(10)

plt.figure(figsize=(10,5))
top_cuisines.plot(kind='bar')
plt.xlabel("Cuisine Combination")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cuisine Combinations")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# Task 2: Cuisine Combinations with Higher Ratings
# -----------------------------

# Average ratings by cuisine combination
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean()

# Top rated cuisine combinations
top_rated = cuisine_ratings.sort_values(ascending=False)

print("\nTop Rated Cuisine Combinations:")
print(top_rated.head(10))
```
