```python
import pandas as pd

# Load Dataset
df = pd.read_csv("Dataset .csv")
```

```python
# Check Price Range Column
print(df['Price range'].head())
```

```python
# Create Bar Chart for Price Range Distribution
price_counts = df['Price range'].value_counts().sort_index()

plt.bar(price_counts.index, price_counts.values)

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Price Ranges")

plt.show()
```

```python
# Calculate Percentage of Restaurants in Each Category
percentage = (price_counts / len(df)) * 100

print(percentage)
```

```python
# Better Output with Labels
for i, value in percentage.items():
    print(f"Price Range {i}: {value:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Dataset .csv")

# Count restaurants in each price range
price_counts = df['Price range'].value_counts().sort_index()

# Bar chart
plt.bar(price_counts.index, price_counts.values)

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Price Range Distribution")

plt.show()

# Percentage calculation
percentage = (price_counts / len(df)) * 100

# Print percentages
for i, value in percentage.items():
    print(f"Price Range {i}: {value:.2f}%")
```

```python
# Histogram
plt.hist(df['Price range'], bins=4)

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Histogram of Price Ranges")

plt.show()
```

```python
# Percentage on Bar Chart
price_counts = df['Price range'].value_counts().sort_index()

bars = plt.bar(price_counts.index, price_counts.values)

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, y,
             ha='center', va='bottom')

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Price Range Distribution")

plt.show()
```
