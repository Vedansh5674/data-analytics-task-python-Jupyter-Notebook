```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

# 1. Most restaurants
city_counts = df['City'].value_counts()
print("Most restaurants:", city_counts.idxmax())

# 2. Average rating per city
avg_rating = df.groupby('City')['Aggregate rating'].mean()

# 3. Best rated city
print("Highest avg rating city:", avg_rating.idxmax())
```

```python
top_5_cities = city_counts.head(5)
print(top_5_cities)
```

```python
top_5_rated_cities = avg_rating.sort_values(ascending=False).head(5)
print(top_5_rated_cities)
```

```python
df = df.dropna(subset=['City', 'Aggregate rating'])
```

```python
df['Aggregate rating'] = df['Aggregate rating'].fillna(df['Aggregate rating'].mean())
```

```python
correlation = df['Aggregate rating'].corr(df['Votes'])
print(f"Correlation between 'Aggregate rating' and 'Votes': {correlation:.2f}")
```

```python
summary = df.groupby('City').agg(
    total_restaurants=('Restaurant Name', 'count'),
    avg_rating=('Aggregate rating', 'mean')
)

print(summary.sort_values(by='avg_rating', ascending=False).head())
```

```python
summary['score'] = summary['total_restaurants'] * summary['avg_rating']
best_city = summary['score'].idxmax()

print("Best overall city:", best_city)
```
