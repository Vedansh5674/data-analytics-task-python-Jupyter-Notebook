```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/Dataset .csv")
```

```python
print(df['Cuisines'].head())
```

```python
# Remove missing values
df = df.dropna(subset=['Cuisines'])

# Split cuisines and create a list
all_cuisines = df['Cuisines'].str.split(', ')
```

```python
from collections import Counter

# Flatten the list
cuisine_list = []

for cuisines in all_cuisines:
    cuisine_list.extend(cuisines)

# Count frequency
cuisine_count = Counter(cuisine_list)
```

```python
top_3 = cuisine_count.most_common(3)

print("Top 3 Cuisines:")
for cuisine, count in top_3:
    print(cuisine, ":", count)
```

```python
total_restaurants = len(df)

print("\nPercentage of Restaurants Serving Top Cuisines:")

for cuisine, count in top_3:
    percentage = (count / total_restaurants) * 100
    print(f"{cuisine}: {percentage:.2f}%")
```

```python
import pandas as pd
from collections import Counter

# Load dataset
df = pd.read_csv("/content/Dataset .csv")

# Remove null values
df = df.dropna(subset=['Cuisines'])

# Split cuisines
all_cuisines = df['Cuisines'].str.split(', ')

# Create cuisine list
cuisine_list = []

for cuisines in all_cuisines:
    cuisine_list.extend(cuisines)

# Count cuisines
cuisine_count = Counter(cuisine_list)

# Top 3 cuisines
top_3 = cuisine_count.most_common(3)

# Total restaurants
total_restaurants = len(df)

# Display results
print("Top 3 Most Common Cuisines:\n")

for cuisine, count in top_3:
    percentage = (count / total_restaurants) * 100

    print(f"Cuisine: {cuisine}")
    print(f"Restaurants: {count}")
    print(f"Percentage: {percentage:.2f}%")
    print()
```
