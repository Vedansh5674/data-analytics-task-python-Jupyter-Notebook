```python
import pandas as pd
import folium

# Load dataset
df = pd.read_csv("Dataset .csv")

# Create base map
restaurant_map = folium.Map(
    location=[df['Latitude'].mean(),
              df['Longitude'].mean()],
    zoom_start=5
)

# Add restaurant locations
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3,
        popup=row['Restaurant Name'],
        fill=True
    ).add_to(restaurant_map)

# Save map
restaurant_map.save("restaurant_map.html")

print("Map has been created successfully!")
```

```python
# To identify clusters/patterns
# Restaurants count by city
city_counts = df['City'].value_counts()

print("Cities with Highest Number of Restaurants:")
print(city_counts.head(10))
```

```python
# clustering visualization
from folium.plugins import MarkerCluster

# Create map
restaurant_map = folium.Map(
    location=[df['Latitude'].mean(),
              df['Longitude'].mean()],
    zoom_start=5
)

# Marker cluster
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Add markers
for index, row in df.iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(marker_cluster)

restaurant_map.save("restaurant_cluster_map.html")
```

```python
# restaurant density by city
print(df['City'].value_counts().head(10))
```

```python
# heatmap visualization
from folium.plugins import HeatMap

heat_data = df[['Latitude', 'Longitude']].dropna().values.tolist()

restaurant_map = folium.Map(
    location=[df['Latitude'].mean(),
              df['Longitude'].mean()],
    zoom_start=5
)

HeatMap(heat_data).add_to(restaurant_map)

restaurant_map.save("restaurant_heatmap.html")
```
